<h2> Data Cleanup </h2>

Import `numpy`, `pandas`, and `matplotlib`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Load in the "states_edu.csv" dataset and take a look at the head of the data

```python
edu = pd.read_csv('../data/states_edu.csv')
edu.columns
```

You should always familiarize yourself with what each column in the dataframe represents. Read about the states_edu dataset here: https://www.kaggle.com/noriuk/us-education-datasets-unification-project

Use this space to rename columns, deal with missing data, etc. _(optional)_

```python
edu.rename({
    'GRADES_PK_G':'ENROLL_PREK',
    'GRADES_KG_G':'ENROLL_K',
    'GRADES_4_G':'4_ENROLL',
    'GRADES_8_G':'8_ENROLL',
    'GRADES_12_G':'12_ENROLL',
    'GRADES_1_8_G':'PRIMARY_ENROLL',
    'GRADES_9_12_G':'HS_ENROLL',
    'GRADES_ALL_G':'TOTAL_ENROLL',
    'ENROLL':'TOTAL_ENROLL_EST'
    },
    axis=1,inplace=True)
edu.head(2000)
```

```python
edu["TOTAL_ENROLL_EST"] = edu["TOTAL_ENROLL_EST"].fillna(edu["TOTAL_ENROLL"])
##define primary enroll
edu["PRIMARY_ENROLL"] = edu["4_ENROLL"] + edu["8_ENROLL"]
##define HS_ENROLL
edu["HS_ENROLL"] = edu["HS_ENROLL"].fillna(edu["12_ENROLL"])
edu.head()
```

```python

##Checking to see what percentage of scores are missing
((edu["AVG_READING_4_SCORE"].notna().sum()/ (edu["AVG_READING_4_SCORE"].isna().sum()+edu["AVG_READING_4_SCORE"].notna().sum()))*100) 
## only ~ 37% of scores filled in
##dropping the missing scores
edu.dropna(subset = ["AVG_READING_4_SCORE"], inplace = True)
edu.shape
```

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
##checking total expenditure is close to sum of expenditure
((edu["TOTAL_EXPENDITURE"]-edu["INSTRUCTION_EXPENDITURE"]-edu["SUPPORT_SERVICES_EXPENDITURE"]-edu["OTHER_EXPENDITURE"]-edu["CAPITAL_OUTLAY_EXPENDITURE"])/edu["TOTAL_EXPENDITURE"]*100).describe()
##They are fairly close mean ~ 4%
## dropping missing expenditure
edu.dropna(subset =["INSTRUCTION_EXPENDITURE"], inplace = True)
##Creating TOTAL EXPENDITURE PER STUDENT
edu['TOTAL_EXPENDITURE_PER_STUDENT'] = edu["TOTAL_EXPENDITURE"] / edu["TOTAL_ENROLL"]
##Creating Instruction Expenditure Per Student
edu['INSTRUCTION_EXPENDITURE_PER_STUDENT'] = edu['INSTRUCTION_EXPENDITURE'] / edu['TOTAL_ENROLL']
##Creating Support Expenditure per Student
edu['SUPPORT_EXPENDITURE_PER_STUDENT'] = edu['SUPPORT_SERVICES_EXPENDITURE'] / edu['TOTAL_ENROLL']
## Support + Instruction exp per student
edu["SI_EXP_PER_STUDENT"] = edu['SUPPORT_EXPENDITURE_PER_STUDENT'] + edu['INSTRUCTION_EXPENDITURE_PER_STUDENT']

edu.count()
```

Feature engineering justification: **< I created 4 collumns for instruction, support, instruction + support, and total expenditure per student. I wanted to look at the correleation between expenditure and student performance.>**

```python
edu.AVG_READING_4_SCORE.plot.hist(title="Frequency of Average 4th Grade Reading Scores", edgecolor="black")
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = edu[["INSTRUCTION_EXPENDITURE_PER_STUDENT","SUPPORT_EXPENDITURE_PER_STUDENT","SI_EXP_PER_STUDENT"]].dropna()
y = edu.loc[X.index]["AVG_READING_4_SCORE"]
y.fillna(y.mean(), inplace=True)
```

```python
##70 / 30 train / test split. 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
score = model.score(X_test, y_test)

mean_error = np.mean(model.predict(X_test)-y_test)
print(score)
print(mean_error)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'INSTRUCTION_EXPENDITURE_PER_STUDENT'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "green")
plt.scatter(X_train[col_name], model.predict(X_train), color = "red")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_READING_4_SCORE')
plt.title("Model Behavior on Training Set")
```

```python
col_name = "INSTRUCTION_EXPENDITURE_PER_STUDENT"

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_READING_4_SCORE')
plt.title("Model Behavior on Testing Set")
```
